```
import codecs
import pandas as pd
import spacy
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```
nlp=spacy.load('/content/drive/My Drive/Colab Notebook/spacy_model')
data=''
corpus = '/content/drive/My Drive/corpus_new'
```

```
import warnings
warnings.filterwarnings('ignore')
```

```
df=pd.read_excel('/content/drive/My Drive/Colab Notebooks/glossary.xls',index_col=0)
```

```
terms = list(df['Term'])
```

```
terms
```

```
corpus = open(corpus,'r')
data = corpus.readlines()

#for line in corpus:
#  data=data+' '+line
```

```
data = " ".join(data)
```

```
data=data.split()
```

```
data[100]
```

```
import nltk
import numpy as np
```

```
doc=[]
for i in terms:
  doc.append(nlp(i))    
```

```
from nltk.probability import FreqDist

u = FreqDist(nltk.ngrams(data,n=1))
u = u.most_common(10000)
#unigram = sorted(set(list(nltk.ngrams(data,n=1))))
print(len(u))
```

```
u[100]
```

```
uni =[]
for i in u:
   j=" ".join(i[0])
   uni.append(nlp(j))
```

```
uni[400]
```

```
synonym_dict={}
for i in terms:
  synonym_dict[i]=[]
```

```
scores = lambda x,doc1:np.array([doc1.similarity(x1) for x1 in x])

def assignment(token,confidence=0.6):
  score = scores(doc,token)
  max_score_ind = np.argmax(score)
  max_score = np.max(score)
  if max_score>confidence:
    synonym_dict[terms[max_score_ind]].append(token.text)
    print(terms[max_score_ind],token.text,max_score)
  
```

```
for i in uni:
  assignment(i)
```

```
del uni
```

```
bigrams = FreqDist(nltk.bigrams(data))
bigrams = bigrams.most_common(4000)
print(len(bigrams))
bi=[]
for i in bigrams:
  j=" ".join(i[0])
  bi.append(nlp(j))
```

```
for i in bi:
  assignment(i)
```

```
del bi
```

```
trigrams = FreqDist(nltk.trigrams(data))
trigrams = trigrams.most_common(4000)
print(len(trigrams))
tri=[]
for i in trigrams:
  j=" ".join(i[0])
  tri.append(nlp(j))
```

```
for i in tri:
  assignment(i)
```

```
import pickel
pickel_out = open('/content/drive/My Drive/data.pickle',"wb")
pickle.dump(pickel_out,synonym_dict)
pickel_out.close()
```
