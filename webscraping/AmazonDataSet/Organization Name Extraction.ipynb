From the extacted product list preparing the list of feature maps which matches closely with the glossary text.

Usually the very first word remains belong to the product name 

for the remaining words following rule based approaches are made use of

1. Sequences and series of noun
2. Adjective if it matches the glossary in the word space. 
3. Adjective followed by noun
4. If the words are not matched then they are to be included in other list of non glossary items
5. ADV followed by noun in glossary terms.
6. ADV DET NOUN combination
7. For the words containing letters alpha numeric comparison to be done using the regular expression (REs). 
8. If verb and in glossary then to be included.

```python
scores = lambda x,doc1:np.array([doc1.similarity(x1) for x1 in x])

def matching_word(token,confidence=0.65):
    if token.vector_norm!=0:
        score = scores(doc_list,token)
        max_score_ind = np.argmax(score)
        max_score = np.max(score)
        if max_score>confidence:
            #print(max_score,doc_list[max_score_ind],token.text)
            return token.text
        else:
            #print("No Match",max_score,doc_list[max_score_ind],token.text)
            return ""
    
    else:
        return ""
```

```python
def getFeatureName(productName):
    productName = productName.lower()
    productName = productName.replace("]","")
    productName = productName.replace("[","")
    productName = productName.replace("/","")
    productName = productName.replace("\\","")
    productName = productName.replace("-","")
    productName = productName.replace("_","")
    productName = productName.replace(",","")
    productName = productName.replace("【","")
    productName = productName.replace("】","")
    norog = ['over','gaming','heavy','foldable','noise','sleep','upgrade','ear','open','3.5mm','bone','new','bulk','bluetooth','in','bluetooth','earbud','wireless','usb','noise','earbuds','headphones','headphones','wholesale','true','earphones']
    prod_features = []
    prod_name = productName
    if len(prod_name.split())>1:
        if prod_name[0] not in norog:
            prod= (" ".join(prod_name.split()[1:])).strip()
        else:
            prod =prod_name
        #print(prod)
        #Pos tagging for getting the features out.  
        doc=nlp1(prod)
        feat = ""
        for token in doc:
            #print(token,token.pos_)
            #Only alphabetic: issue
            if pattern.match(token.text):
                #print(token.text)
                if len(token.text)==1:
                    pass
                #print(token.text,token.pos_)
                elif token.pos_ in ('NOUN','ADJ','ADV') and token.text not in non_nouns:
                    if (token.text in numerics) and doc[token.i+1].text=="driver":
                        feat+=token.text+" "
                        #prod_features.append(token.text+" "+doc[token.i+1].text)
                        #productName=productName.replace(token.text+" "+doc[token.i+1].text,"")
                    else:
                        if token.pos_ in ('ADJ','ADV'):
                            if feat!="":
                                feat +=token.text+" " 
                                feat_doc=nlp1(feat.strip())
                                wm = matching_word(feat_doc)
                                if wm!="":
                                    prod_features.append(wm)
                                    productName=productName.replace(wm,"")
                                    feat=""
                                #Indicating if the token has a match but not the entire document 
                                else:
                                    wm = matching_word(token)
                                    if wm!="":
                                        prod_features.append(wm)
                                        productName=productName.replace(wm,"")
                                        #feat does not map here thus adding the feat to  non existing glossary
                                        prod_features.append(feat)
                                        productName=productName.replace(feat,"")
                                        words_not_in_glosssary.append(feat)
                                        feat=""
                            elif token.text in adj:
                                prod_features.append(token.text)
                                productName=productName.replace(token.text,"")
                                words_not_in_glosssary.append(token.text)
                            elif token.text in no_adj:
                                pass
                            else:
                                wm = matching_word(token)
                                if wm!="":
                                    prod_features.append(wm)
                                    productName=productName.replace(wm,"")
                                else:
                                    feat+=token.text+" "
                        else: #Word considered is a noun
                            if feat!="": 
                                feat +=token.text+" "
                                feat_doc=nlp1(feat.strip())
                                wm = matching_word(feat_doc)
                                if wm!="":
                                    prod_features.append(wm)
                                    productName=productName.replace(wm,"")
                                    feat=""
                                #Indicating if the token has a match but not the entire document 
                                else:
                                    wm = matching_word(token)
                                    if wm!="":
                                        prod_features.append(wm)
                                        productName=productName.replace(wm,"")
                                        #feat does not map here thus adding the feat to  non existing glossary
                                        feat=feat.replace(token.text,"").strip()
                                        prod_features.append(feat)
                                        productName=productName.replace(feat,"")
                                        words_not_in_glosssary.append(feat)
                                        feat=""
                                    else:
                                        prod_features.append(feat.strip())
                                        productName=productName.replace(feat.strip(),"")
                                        words_not_in_glosssary.append(feat.strip())
                                        feat=""
                            else:
                                if token.text == "bone" and (doc[token.i+1].text=="conductor" or doc[token.i+1].text=="conduction"): 
                                    words_not_in_glosssary.append(token.text+" "+doc[token.i+1].text)
                                    prod_features.append(token.text+" "+doc[token.i+1].text)
                                    productName=productName.replace(token.text+" "+doc[token.i+1].text,"")
                                else:
                                    wm = matching_word(token)
                                    if wm=="":
                                        words_not_in_glosssary.append(token.text)
                                    prod_features.append(token.text)
                                    productName=productName.replace(token.text,"")
                elif (token.pos_ =="ADP" or token.pos_=="PART") and (feat=="" or token.text=="in") and matching_word(token,confidence=0.6):
                    #print(token.text)
                    if token.text in ["in",'on']:
                        if doc[token.i+1].text=="ear":
                            feat+=token.text+" "

                        
                    else:
                        feat+=token.text+" "
                    #print(feat)
                elif (token.pos_ =="ADP" or token.pos_=="PART")  and feat=="":
                    productName=productName.replace(token.text,"")  
                elif token.pos_=="DET" and feat!="":
                    feat+=token.text+" "
                elif token.pos_ == "VERB":
                    if token.text in verb_listing:
                        feat+=doc[token.i-1].text+" "+token.text
                        wm = matching_word(nlp1(feat))
                        if wm=="":
                            words_not_in_glosssary.append(feat)
                        prod_features.append(feat)
                        productName=productName.replace(feat,"")
                        feat=""
                        
                    elif (token.text =="closed" or token.text=="open") or (token.i<len(doc)-1 and doc[token.i+1].text)=="back":
                        if(token.i<len(doc)-1):
                            prod_features.append(token.text+" "+doc[token.i+1].text)
                            words_not_in_glosssary.append(token.text+" "+doc[token.i+1].text)
                            productName=productName.replace(token.text+" "+doc[token.i+1].text,"")
                elif token.pos_=="ADV" and token.text in adv:
                    words_not_in_glosssary.append(token.text)
                    prod_features.append(token.text)
                    productName=productName.replace(token.text,"") 
                elif token.pos_=="CCONJ" and token.text=="and":
                      productName=productName.replace(token.text,"")    
            elif token.pos_ =="NUM" and numbers.match(token.text) and token.i<(len(doc)-1) and doc[token.i+1].text=="driver":    
                feat+=token.text+" "
            elif token.text in colors:
                    prod_features.append(token.text)
                    productName = productName.replace(token.text,"")
        if feat!="":
            prod_features.append(feat)
            productName= productName.replace(feat.strip(),"")
            words_not_in_glosssary.append(feat)
            feat=""
    for i in feature_list:
        if i in productName:
            productName=productName.replace(i,"")
            prod_features.append(i)
    
    prod_features = ",".join(prod_features)
    product_feature.append(prod_features)
    productName=re.sub("[ ]+"," ",productName)
    #productName=productName.replace("  "," ")
    return (productName,prod_features)
```
